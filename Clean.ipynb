```python
import pandas as pd
import numpy as np
```

# Dataset 1 (Partidos entre 2020-2022)

```python
WR = pd.read_csv("data/W Race Record.csv", index_col='Race', encoding="utf-16", sep= "\t")

WR
```

```python
WR['TDs Net'] = WR['TDs Net'].str.replace(',00','') #Quitar los ,00 del "float" para poder pasarlos a float reales
WR['# Losses'] = WR['# Losses'].str.replace(',00','')
```

```python
WR['Prom. Win %'] = WR['Prom. Win %'].map(lambda x: x.rstrip('%')) #Quitar los % y cambiar la coma a punto para poder operar más adelante.
WR['Prom. Win %'] = WR['Prom. Win %'].str.replace(',','.')
WR['Prom. Win %'] = WR['Prom. Win %'].astype(float)
```

```python
WR['# Losses'] = WR['# Losses'].astype(float)
WR['TDs Net'] = WR['TDs Net'].astype(float)
```

# Dataset 2 (Eurobowl22)

```python
WR2 = pd.read_csv("data/W Race Record(1).csv", index_col='Race', encoding="utf-16", sep= "\t")

WR2
```

```python
WR2['# Games'] = WR2['# Games'].str.replace(',00','')
WR2['# Wins'] = WR2['# Wins'].str.replace(',00','')
WR2['# Draws'] = WR2['# Draws'].str.replace(',00','')
WR2['# Losses'] = WR2['# Losses'].str.replace(',00','')
WR2['TDs For'] = WR2['TDs For'].str.replace(',00','')
WR2['TDs Against'] = WR2['TDs Against'].str.replace(',00','')
WR2['TDs Net'] = WR2['TDs Net'].str.replace(',00','')
WR2['Prom. TDs For'] = WR2['Prom. TDs For'].str.replace(',','.')
WR2['Prom. TDs Against'] = WR2['Prom. TDs Against'].str.replace(',','.')
#lo mismo que antes, quitar las comas de los strings para poder pasarlos a float
```

```python
WR2['Prom. Win %'] = WR2['Prom. Win %'].map(lambda x: x.rstrip('%')) #adaptar la columna de % para pasarlo a float
WR2['Prom. Win %'] = WR2['Prom. Win %'].str.replace(',','.')
WR2['Prom. Win %'] = WR2['Prom. Win %'].astype(float)
```

```python
WR2['# Games'] = WR2['# Games'].astype(float)
WR2['# Wins'] = WR2['# Wins'].astype(float)
WR2['# Draws'] = WR2['# Draws'].astype(float)
WR2['# Losses'] = WR2['# Losses'].astype(float)
WR2['TDs Net'] = WR2['TDs Net'].astype(float)
WR2['TDs For'] = WR2['TDs For'].astype(float)
WR2['TDs Against'] = WR2['TDs Against'].astype(float)
WR2['Prom. TDs For'] = WR2['Prom. TDs For'].astype(float)
WR2['Prom. TDs Against'] = WR2['Prom. TDs Against'].astype(float)
#cambio de str a float de todas las columnas
```

# Unión

```python
WR3 = WR #creación de "copias" para no tener que volver a rehacer todo lo anterior
hastadosmilveintidos = WR
dosmilveintidos = WR2
```

```python
WRdef = pd.concat([WR3, WR2]).groupby(['Race']).sum().reset_index() #unión de los dos datasets

WRdef
```

```python
WRdef["Prom. Win %"] = WRdef["Prom. Win %"] / 2 # como solo hemos sumado en la columna del porcentaje de victorias, hay que dividirlo entre el número de datasets

WRdef
```

```python
WRfinal = WRdef.drop([2, 6, 17, 29]) # quito 4 filas de equipos/total general no relevantes
WRfinal.set_index('Race')
```

```python
WRfinal.sort_values(by='Prom. Win %', ascending=False) #para ver qué raza tiene mayor porcentaje de victorias
```

```python
WRfinal.sort_values(by='# Games') #número de partidos jugados, del menos popular al que más
```

```python
WRfinal.to_csv('ESTADISTICAS.csv', index=False) #creación del archivo csv final
```

```python
WRfinal.to_excel('ESTADISTICAS.xlsx', index=False)
```
