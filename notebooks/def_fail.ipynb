```python
def sum_csv_columns(folder_path, column_name):
  # Obtener la lista de archivos .csv de la carpeta
  all_filenames = glob.glob(folder_path + "/*.csv")

  # Inicializar un diccionario para guardar los valores de la columna
  column_values = {}

  # Iterar sobre la lista de archivos .csv
  for file in all_filenames:
      # Leer el archivo .csv
      df = pd.read_csv(file, encoding="utf-16", sep= "\t")
      # Obtener los valores de la columna y sumarlos
      values = df[column_name].values
      column_values[file] = values.sum()

  # Crear un dataframe a partir del diccionario
  result_df = pd.DataFrame.from_dict(column_values, orient='index', columns=[column_name])

  # Guardar el dataframe resultante en un .csv
  result_df.to_csv("summed_csv.csv", index=True, encoding='utf-8-sig')


sum_csv_columns("C:\\Users\\PCCOM\\Desktop\\TheBridge\\BB_CSV\\teams", "Number of Games (Tournament) (copy)")
```

```python
def merge_csv_files(folder_path, output_file):
  # Obtener la lista de archivos .csv de la carpeta
  all_filenames = glob.glob(folder_path + "/*.csv")

  # Inicializar un diccionario para guardar los valores de las columnas
  column_values = {"Race": [], "Number of Games (Tournament) (copy)": 0, "Prom. Win %": 0}

  # Iterar sobre la lista de archivos .csv
  for file in all_filenames:
      # Leer el archivo .csv
      df = pd.read_csv(file, encoding="utf-16", sep= "\t")
      # Obtener los valores de las columnas y sumarlos
      column_values["Race"] += df["Race"].values.tolist()
      column_values["Number of Games (Tournament) (copy)"] += df["Number of Games (Tournament) (copy)"].sum()
      # Pasar la columna "Prom. Win %" a tipo int
      df["Prom. Win %"] = pd.to_numeric(df["Prom. Win %"], errors='coerce')
      column_values["Prom. Win %"] += df["Prom. Win %"].sum()

  # Crear un dataframe a partir del diccionario
  result_df = pd.DataFrame.from_dict(column_values)

  # Guardar el dataframe resultante en un .csv
  result_df.to_csv(output_file, index=False, encoding='utf-8-sig')


merge_csv_files("C:\\Users\\PCCOM\\Desktop\\TheBridge\\BB_CSV\\teams", "merged_csv.csv")
```
